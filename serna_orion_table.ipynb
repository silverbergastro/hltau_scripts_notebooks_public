# serna_orion_table

Read in table from Serna et al. (2021) and evaluate to identify longest periods for classical T Tauri stars in Orion.

```python
from astropy.table import Table
import matplotlib.pyplot as plt
```

```python
testtable = Table.read('hltau_tables/apjac300at1_mrt.txt',format='mrt')
testtable.show_in_notebook()
```

```python
classmask = testtable['AClass'] == 'C'

testtableCTTS = testtable[classmask]
testtable[classmask].show_in_notebook()
```

```python
print([i for i in range(len(testtable['Per'])) if testtable['Per'][i] > 8.])
```

```python
mask = testtable['Per'] > 8.

testtable[mask].show_in_notebook()
```

```python
plt.figure()
plt.hist(testtableCTTS['Per'])
plt.xlabel('Period (days)')
plt.show()
```
